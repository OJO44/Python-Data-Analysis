**Investors flow of funds**

```python
#import the necessary libraries
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
```

```python
#tep 2. Import the dataset from this address.
#Step 3. Assign it to a variable called Investors
data = "https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv"
investors = pd.read_csv(data)
investors.head()
```

```python
#Step 4. What is the frequency of the dataset?
investors.describe()
```

```python
#Step 5. Set the column Date as the index.
investors.set_index("Date", inplace=True)
investors.head()
```

```python
#Step 6. What is the type of the index?
#investors.index.dtype
print(type(investors.index))
```

```python
#Step 7. Set the index to a DatetimeIndex type
investors.index = pd.DatetimeIndex(investors.index)
print(investors.index)
```

```python
#Step 8. Change the frequency to monthly, sum the values and assign it to monthly.
monthly= investors.resample("M").sum()
print(monthly)
```

```python
#Step 9. You will notice that it filled the dataFrame with months that don't have any data with NaN. Let's drop these rows.
monthly = monthly.dropna()
print(monthly)
```

```python
#Step 10. Good, now we have the monthly data. Now change the frequency to year.
yearly = monthly.resample('Y').sum()
print(yearly)
```
