**Deleting**

```python
#Step 1. Import the necessary libraries
import pandas as pd
```

```python
#Step 2. Import the dataset from this address.
#Step 3. Assign it to a variable called iris
data = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
Iris = pd.read_csv(data)
Iris.head()
```

```python
#Step 4. Create columns for the dataset
# 1. sepal_length (in cm)
# 2. sepal_width (in cm)
# 3. petal_length (in cm)
# 4. petal_width (in cm)
# 5. class
column_names = ["sepal_length_cm","sepal_width_cm","petal_length_cm","petal_width_cm","class"]
Iris= pd.read_csv(data, names = column_names)
Iris.head()
```

```python
#Step 5. Is there any missing value in the dataframe?
missing_values = Iris.isnull().sum()
print(missing_values)
```

```python
#Step 6. Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN
Iris.loc[10:29, 'petal_length_cm'] = float('NaN')
Iris.head(30)
```

```python
#Step 7. Good, now lets substitute the NaN values to 1.0
Iris["petal_length_cm"].fillna(1.0,inplace= True)
Iris.head(30)
```

```python
#Step 8. Now let's delete the column class
Iris.drop("class",axis=1, inplace= True)
Iris.head()
```

```python
#Step 9. Set the first 3 rows as NaN
Iris.iloc[:3,:] = float("NaN")
Iris.head()
```

```python
#Step 10. Delete the rows that have NaN
Iris.dropna(axis=0, how= "any", inplace=True)
Iris.head()
```

```python
#Step 11. Reset the index so it begins with 0 again
Iris.reset_index(drop=True, inplace= True)
Iris.head()
```
