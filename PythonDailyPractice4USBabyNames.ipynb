**US Baby Names**

```python
#Step1 Import the necessary Libraries
import pandas as pd
import numpy as np

```

```python
#Step 2. Import the dataset from this address.
#Step 3. Assign it to a variable called baby_names.
#Step 4. See the first 10 entries
data = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv"
baby_names = pd.read_csv(data)
baby_names.head(10)
```

```python
#Step 5. Delete the column 'Unnamed: 0' and 'Id'
baby_names.drop(columns= ["Unnamed: 0", "Id"])
```

```python
#Step 6. Is there more male or female names in the dataset?
gender_counts = baby_names['Gender'].value_counts()
print(gender_counts)
#No there are more females than males
```

```python
#Step 7. Group the dataset by name and assign to names
names= baby_names.groupby("Name")
print(names)
```

```python
#Step 8. How many different names exist in the dataset?
uniqe_names= baby_names["Name"].nunique()
print(uniqe_names)
```

```python
#Step 9. What is the name with most occurrences?
name_with_most_occurence = baby_names["Name"].value_counts().idxmax()
print(name_with_most_occurence)
```

```python
#Step 10. How many different names have the least occurrences?
least_occurence_names = baby_names["Name"].value_counts().min().sum()
print(least_occurence_names)
```

```python
#Step 11. What is the median name occurrence?
median_name_occurence = baby_names["Name"].value_counts().median()
print(median_name_occurence)
```

```python
#Step 12. What is the standard deviation of names?
name_std = baby_names["Name"].value_counts().std()
print(name_std)
```

```python
#Step 13. Get a summary with the mean, min, max, std and quartiles
names_summary = baby_names['Name'].value_counts().describe()
print(names_summary)
```
